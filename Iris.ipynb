Load the data in the file Iris.txt in a 2-D array called
iris.

```python
# case study 1
import numpy as np
iris=np.genfromtxt("Iris.txt",delimiter=",")
print(iris)
```

Drop column whose index = 4 from the array iris.

```python
iris_new=np.delete(iris,4,axis=1)# for columns
print(iris_new)
```

Display the shape, dimensions and size of iris.

```python
print(np.ndim(iris_new))
print(np.shape(iris_new))
print(np.size(iris_new))
```

 Split iris into three 2-D arrays, each array for a different
species. Call them irisl, iris2, iris3.

```python
iris1,iris2,iris3=np.split(iris_new,[10,20])
```

```python
print(iris1)
```

Find the max, min, mean and standard deviation for the
columns of the iris and store the results in the arrays
iris_max, iris_min, iris_avg, iris_std, iris
var respectively. The results must be rounded to not
more than two decimal places.

```python
iris_max=np.round(iris_new.max(axis=0),2)# column wise
print(iris_max)
iris_min=np.round(iris_new.min(axis=0),2)# column wise
print(iris_min)
iris_avg=np.round(iris_new.mean(axis=0),2)# column wise
print(iris_avg)
iris_median=np.round(np.median(iris_new,axis=0),2)# column wise
print(iris_median)
iris_var=np.round(np.var(iris_new,axis=0),2) # column wise
print(iris_var)
iris_std=np.round(np.std(iris_new,axis=0),2)# column wise
print(iris_std)
```
